```python
from numba import jit
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
import time
from scipy import interpolate
from tqdm.auto import tqdm
```

```python
## Setting Initial Guess ##
def getInitialGuess(parameters):
    Nx = parameters[0]
    if len(parameters) >= 5:
        Ny = parameters[4]
    else:
        Ny=Nx
    Nh = parameters[1]
    Nl = parameters[2]
    initialGuess = np.resize(np.zeros(1), (Nx+1, Ny+1))
    for i in range(Nl+1):
        initialGuess[i][-Nh-1] = 0.5*V0
    return initialGuess

## Convergence Criteria: Frobenius Norm ##
@jit(nopython=True)
def frobeniusNorm(vector):
    summation = 0
    for row in vector:
        for el in row:
            summation += el**2
    norm = (summation)**0.5
    return norm    

## SOR Method for Test Case (1st quadrant) ##

@jit(nopython=True)
def sorMethod2D(potential, parameters, w=1, convergenceCriteria=1e-4):
    residue = 1
    iterations = 0
    Nx = parameters[0]
    if len(parameters) >= 5:
        Ny = parameters[4]
    else:
        Ny=Nx
    Nh = parameters[1]
    Nl = parameters[2]
    while residue > convergenceCriteria:
        iterations += 1
        formerPotential = np.copy(potential)
        for i in range(Nx):
            for j in range(1, Ny):
                if j == Ny-Nh and i <= Nl:
                    continue
                if i==0:
                    potential[i][j] = (1-w)*potential[i][j] + w*(potential[i][j-1] + potential[i][j+1] + potential[i+1][j] + potential[i][j])/4
                    continue
                potential[i][j] = (1-w)*potential[i][j] + w*(potential[i][j-1] + potential[i][j+1] + potential[i+1][j] + potential[i-1][j])/4
        residue = frobeniusNorm(potential - formerPotential)
        if iterations >= 50000:
            break
    return potential, iterations
```

```python
## Test Case ##
V0 = 1
lx = 2
d = 2
Dx, Dy = 2, 2
h = 0.5
Nx, Ny, Nh, Nl = int(Dx/h), int(Dy/h), int(d/(2*h)), int(lx/(2*h))
parameters = (Nx, Nh, Nl, Dx, Ny, Dy)
initialGuess = getInitialGuess(parameters)
potential, iterations = sorMethod2D(initialGuess, parameters, convergenceCriteria = 1e-2)

print(f'Iterations: {iterations}')
fig, ax = plt.subplots(1, 1)
im = ax.imshow(np.transpose(potential), cmap='jet')
fig.colorbar(im, ticks=[-0.5, -0.25, 0, 0.25, 0.5])

print(f'Test Cases:')
print(f'V[0,1] = {potential[0][1]:.3f}, V[0,3] = {potential[0][3]:.3f}')
print(f'V[1,1] = {potential[1][1]:.3f}, V[1,3] = {potential[1][3]:.3f}')
print(f'V[2,1] = {potential[2][1]:.3f}, V[2,3] = {potential[2][3]:.3f}')
print(f'V[3,1] = {potential[3][1]:.3f}, V[3,3] = {potential[3][3]:.3f}')
print(f'V[3,2] = {potential[3][2]:.3f}')
print(f'Observation: Convergence Criteria was set to 1e-2 to get desired values')
```

```python
## Potential on XY Plane ##

def potentialOnPlane(potential):
    quarterPotential = np.transpose(potential)
    totalPotential = np.resize(np.zeros(1), (2*Ny+1, 2*Nx+1))
    for i in range(Ny+1):
        for j in range(Nx+1):
            totalPotential[i][-j] = quarterPotential[i][-j]
            totalPotential[i][j] = quarterPotential[i][-j-1]
            totalPotential[-i-1][j] = -totalPotential[i][j]
            totalPotential[-i-1][-j-1] = totalPotential[-i-1][j]
    return totalPotential

def getPotential(parameters, convergenceCriteria=1e-5, w=1):
    initialGuess = getInitialGuess(parameters)
    quadrantPotential, iterations = sorMethod2D(initialGuess, parameters, convergenceCriteria=convergenceCriteria)
    totalPotential = potentialOnPlane(quadrantPotential)
    return totalPotential, iterations

## Plot Formatation ##
def plotPotentialFormat(totalPotential, Dx, Dy=Dx, width=7, height=7):
    fig, ax = plt.subplots()
    fig.set_figwidth(width)
    fig.set_figheight(height)
    im = ax.imshow(totalPotential, cmap='jet',extent=[-Dx, Dx, -Dy, Dy])
    plt.title('Electric Potential Distribution [V]', fontsize=16)
    plt.xlabel('X-axis [length]', fontsize=14)
    plt.ylabel('Y-axis [length]', fontsize=14)
    fig.colorbar(im, ticks=[-0.5, -0.25, 0, 0.25, 0.5], fraction=0.046, pad=0.04)

## Test Case ##
V0 = 1
lx = 3
d = 2
Dx, Dy = 4, 2
h = 0.5
Nx, Ny, Nh, Nl = int(Dx/h), int(Dy/h), int(d/(2*h)), int(lx/(2*h))
parameters = (Nx, Nh, Nl, Dx, Ny, Dy)

totalPotential, iterations = getPotential(parameters)
plotPotentialFormat(totalPotential, Dx, Dy, width=12)
```

```python
def chargePerLength(electricField, parameters):
    n1 = (0, 1)
    n2 = (1, 0)
    n3 = (0, -1)
    n4 = (-1, 0)
    chargePerLength = 0
    for i in range(Nx-Nl-1, Nx+Nl+1):
        el1 = electricField[1][Ny-Nh-1][i]*n1[0]+electricField[0][Ny-Nh-1][i]*n1[1]
        el3 = electricField[1][Ny-Nh+1][i]*n3[0]+electricField[0][Ny-Nh+1][i]*n3[1]
        chargePerLength += (el1 + el3)
    for i in range(Ny-Nh-1, Ny-Nh+1):
        el2 = electricField[1][i][Nx+Nl+1]*n2[0]+electricField[0][i][Nx+Nl+1]*n2[1]
        el4 = electricField[1][i][Nx-Nl-1]*n4[0]+electricField[0][i][Nx-Nl-1]*n4[1]
        chargePerLength += (el1 + el3)
    return chargePerLength

charge = chargePerLength(electricField, parameters)
charge
```
